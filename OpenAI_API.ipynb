```python
from openai import OpenAI
import os
OPENAI_API_KEY='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'
client = OpenAI(api_key=OPENAI_API_KEY)
model = "gpt-3.5-turbo-instruct" 

def chat():
    system = ("Here's a user's conversation with a versatile and "
              f"friendly AI from OpenAI named {model} using GPT-3:\n")
    chat = ""
    stop = "\nuser:"
    user = "Hi! Write an introduction telling me your AI capabilities."
    prompt = "\nAI:"
    while user and user != "exit":
        completion = client.completions.create(
            model=model, max_tokens=1000, stream=True,
            stop=stop, top_p=0.9,
            prompt=(system + chat[-3000:] + stop + user + prompt))

        chat += (stop + user + prompt + (print_streaming_response(completion) or "").strip())
        user = input("\nuser: ").strip()


def print_streaming_response(response, line_length=60):
    reply = ""
    line = ""
    print_started = False
    print(f"{model}:")

    for delta in response:
        if not delta.choices[0].finish_reason:
            content = delta.choices[0].text or ""
            reply += content

            for char in content:
                if not print_started and not char.isspace():
                    print_started = True

                if print_started:
                    if char == '\n':
                        if line:
                            print(line)
                            line = ""
                    else:
                        line += char
                        if len(line) >= line_length:
                            last_space = line.rfind(' ', 0, line_length)
                            if last_space > -1:
                                print(line[:last_space])
                                line = line[last_space + 1:]
                            else:
                                print(line)
                                line = ""
    if line:
        print(line)
    return reply.strip()

chat()
```
